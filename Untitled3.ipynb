```python
import nltk
from newspaper import Article
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('brown')
```

```python
url = 'https://qrius.com/are-middle-lanes-fastest-in-track-and-field/'
```

```python
article = Article(url, language="en") # en for English 
```

```python
article.download() 
article.parse() 
article.nlp() 
```

```python
print("Article Title:") 
print(article.title) #prints the title of the article
```

```python
print("Article Keywords:")
print(article.keywords) #prints the keywords of the article
```

```python
print("Article Text:") 
print(article.text) #prints the entire text of the article
```

```python
txt = article.text
```

```python
import tweepy
from textblob import TextBlob
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
allwords = ''.join([twts for twts in txt])
wordcloud = WordCloud(width=500,height=300, random_state=21,  max_font_size=119).generate(allwords)
plt.imshow(wordcloud, interpolation="bilinear")
```
